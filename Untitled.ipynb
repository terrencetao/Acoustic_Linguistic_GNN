```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA


# === Chargement Dataset ===
def load_data(data_path, batch_size=16):
    train_ds, _ = tf.keras.utils.audio_dataset_from_directory(
        directory=data_path,
        batch_size=batch_size,
        output_sequence_length=16000,
        validation_split=0.2,
        seed=0,
        subset='both'
    )
    return train_ds, train_ds.class_names


# === Embedding TRILL ===
def extract_trill_embeddings(dataset):
    trill_model = hub.load("https://tfhub.dev/google/nonsemantic-speech-benchmark/trill/3")
    embeddings, labels = [], []

    for batch in dataset:
        waveforms, batch_labels = batch
        for waveform, label in zip(waveforms, batch_labels):
            waveform = tf.squeeze(waveform)
            waveform = tf.cast(waveform, tf.float32)
            waveform = waveform / tf.reduce_max(tf.abs(waveform))  # normalize
            waveform = tf.reshape(waveform, [1, -1])  # [1, time]
            embedding = trill_model(waveform, sample_rate=16000)['embedding']  # [1, T, D]
            embedding = tf.reduce_mean(embedding, axis=1)  # [1, D]
            embeddings.append(tf.squeeze(embedding).numpy())
            labels.append(label.numpy())

    return np.array(embeddings), np.array(labels)


# === Embedding Wav2Vec2.0 ===
def pad_or_trim(audio, target_len=246000):
    audio_len = tf.shape(audio)[-1]
    if audio_len < target_len:
        pad_len = target_len - audio_len
        audio = tf.pad(audio, [[0, 0], [0, pad_len]])
    else:
        audio = audio[:, :target_len]
    return audio
def extract_wav2vec_embeddings(dataset):
    wav2vec_handle = "https://tfhub.dev/vasudevgupta7/wav2vec2/1"
    wav2vec_model = hub.KerasLayer(wav2vec_handle, trainable=False)
    embeddings, labels = [], []

    for batch in dataset:
        waveforms, batch_labels = batch
        for waveform, label in zip(waveforms, batch_labels):
            waveform = tf.squeeze(waveform)
            waveform = tf.cast(waveform, tf.float32) / 32768.0  # normalize
            waveform = tf.reshape(waveform, [1, -1])
            waveform = pad_or_trim(waveform, 246000)
            embedding = wav2vec_model(waveform)['contextualized']
            embedding = tf.reduce_mean(embedding, axis=1)
            embeddings.append(tf.squeeze(embedding).numpy())
            labels.append(label.numpy())

    return np.array(embeddings), np.array(labels)
# === Visualisation des embeddings ===
def plot_embeddings(embeddings, labels, label_names, title='Embeddings'):
    if embeddings.shape[1] > 50:
        print("Using t-SNE for dimensionality reduction...")
        reducer = TSNE(n_components=2, perplexity=20, random_state=0)
    else:
        reducer = PCA(n_components=2)

    reduced = reducer.fit_transform(embeddings)
    plt.figure(figsize=(10, 6))
    for i, label_name in enumerate(label_names):
        idxs = labels == i
        plt.scatter(reduced[idxs, 0], reduced[idxs, 1], label=label_name, alpha=0.6)
    plt.legend()
    plt.title(title)
    plt.grid(True)
    plt.tight_layout()
    plt.show()


```

```python
# === Main Script ===
if __name__ == "__main__":
    # Chemin vers ton dossier audio organisé par dossier de classes
    data_path = "data/mini_speech_commands"

    # 1. Charger les données
    dataset, label_names = load_data(data_path)

    # 2. Extraire les embeddings
    print("→ Extraction TRILL...")
    trill_embeds, trill_labels = extract_trill_embeddings(dataset)

    
    # 3. Visualiser
    print("→ Visualisation TRILL...")
    plot_embeddings(trill_embeds, trill_labels, label_names, title="TRILL Embeddings")

    
```

```python
import torch
import torchaudio
import librosa
import numpy as np
from transformers import Wav2Vec2Processor, Wav2Vec2Model
import os
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Charger le modèle Wav2Vec2
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base")
model = Wav2Vec2Model.from_pretrained("facebook/wav2vec2-base")

# Désactiver le mode entraînement
model.eval()

def extract_wav2vec_embedding(audio_path):
    # Charger l'audio
    waveform, sr = librosa.load(audio_path, sr=16000)  # Resample à 16 kHz
    inputs = processor(waveform, sampling_rate=16000, return_tensors="pt", padding=True)

    # Extraire les embeddings
    with torch.no_grad():
        outputs = model(**inputs)

    hidden_states = outputs.last_hidden_state  # [batch_size, time_steps, hidden_size]
    mean_embedding = hidden_states.mean(dim=1).squeeze().numpy()  # Moyenne temporelle

    return mean_embedding
```

```python
import glob

data_folder = "data/mini_speech_commands"
all_embeddings = []
all_labels = []

for class_folder in os.listdir(data_folder):
    class_path = os.path.join(data_folder, class_folder)
    if os.path.isdir(class_path):
        for file_path in glob.glob(f"{class_path}/*.wav"):
            try:
                emb = extract_wav2vec_embedding(file_path)
                all_embeddings.append(emb)
                all_labels.append(class_folder)
            except Exception as e:
                print(f"Erreur pour {file_path}: {e}")
```

```python
from sklearn.preprocessing import LabelEncoder

# Encodage des labels
le = LabelEncoder()
y_encoded = le.fit_transform(all_labels)

# t-SNE
tsne = TSNE(n_components=2, random_state=42)
reduced = tsne.fit_transform(np.array(all_embeddings))

# Affichage
plt.figure(figsize=(10, 8))
scatter = plt.scatter(reduced[:, 0], reduced[:, 1], c=y_encoded, cmap='tab10')
plt.legend(handles=scatter.legend_elements()[0], labels=le.classes_)
plt.title("t-SNE des embeddings Wav2Vec2")
plt.show()
```
