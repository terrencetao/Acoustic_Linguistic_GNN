```python
import torch
import dgl
import pickle
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.metrics.pairwise import pairwise_distances
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import networkx as nx
from collections import defaultdict
import os
from gnn_heto_link_pred_model import HeteroLinkGCN

# Set up plotting style
plt.style.use('seaborn')
sns.set_palette("husl")
```

```python
# 1. Load the Heterogeneous Graph
def load_heterograph(input_folder, graph_file):
    """Load the saved heterograph"""
    glist, _ = dgl.load_graphs(os.path.join(input_folder, graph_file))
    hetero_graph = glist[0]
    
    print("Graph loaded with node types:", hetero_graph.ntypes)
    print("Acoustic nodes:", hetero_graph.num_nodes('acoustic'))
    print("Word nodes:", hetero_graph.num_nodes('word'))
    
    return hetero_graph
```

```python




# Example usage:
# hetero_graph = load_heterograph("your_input_folder", "graph_file.bin")

# 2. Extract Initial Embeddings
def get_initial_embeddings(hetero_graph):
    """Extract the initial feature embeddings"""
    features = {
        'acoustic': hetero_graph.nodes['acoustic'].data['feat'],
        'word': hetero_graph.nodes['word'].data['feat']
    }
    
    # Get phoneme labels if available
    phoneme_labels = hetero_graph.nodes['acoustic'].data.get('label', None)
    
    return features, phoneme_labels

# 3. Load Trained Model and Extract Learned Embeddings
def load_model_and_get_embeddings(model_path, hetero_graph, in_feats):
    """Load trained model and extract embeddings"""
    # Initialize model (adjust parameters as needed)
    hidden_size = 512
    linear_hidden_size = 64
    nb_phon = hetero_graph.nodes['word'].data['feat'].shape[1]
    
    model = HeteroLinkGCN(in_feats, hidden_size, nb_phon, linear_hidden_size)
    model.load_state_dict(torch.load(model_path))
    model.eval()
    
    # Get embeddings
    with torch.no_grad():
        features = {
            'acoustic': hetero_graph.nodes['acoustic'].data['feat'],
            'word': hetero_graph.nodes['word'].data['feat']
        }
        embeddings = model(hetero_graph, features)
    
    return embeddings['acoustic']

# 4. Visualization Functions
def plot_embeddings(embeddings, labels=None, title="Embeddings"):
    """Visualize embeddings using t-SNE"""
    tsne = TSNE(n_components=2, random_state=42)
    emb_2d = tsne.fit_transform(embeddings.cpu().numpy())
    
    plt.figure(figsize=(10, 8))
    if labels is not None:
        # Convert labels to numeric if they aren't already
        if isinstance(labels[0], str):
            le = LabelEncoder()
            numeric_labels = le.fit_transform(labels.cpu().numpy())
        else:
            numeric_labels = labels.cpu().numpy()
            
        scatter = plt.scatter(emb_2d[:, 0], emb_2d[:, 1], c=numeric_labels, 
                            cmap='viridis', alpha=0.7)
        plt.colorbar(scatter, label='Word Classes')
    else:
        plt.scatter(emb_2d[:, 0], emb_2d[:, 1], alpha=0.7)
    
    plt.title(title)
    plt.xlabel("t-SNE 1")
    plt.ylabel("t-SNE 2")
    plt.show()

# 5. Inertia Calculation (Within-cluster variance)
def calculate_inertia(embeddings, labels):
    """Calculate inertia for each phoneme cluster"""
    if labels is None:
        raise ValueError("Labels are required for inertia calculation")
    
    embeddings_np = embeddings.cpu().numpy()
    labels_np = labels.cpu().numpy()
    
    unique_labels = np.unique(labels_np)
    inertia_results = {}
    
    for label in unique_labels:
        cluster_points = embeddings_np[labels_np == label]
        if len(cluster_points) > 1:  # Need at least 2 points to compute variance
            centroid = np.mean(cluster_points, axis=0)
            distances = np.sum((cluster_points - centroid) ** 2, axis=1)
            inertia = np.sum(distances)
            inertia_results[label] = inertia
        else:
            inertia_results[label] = 0.0
    
    # Also calculate global inertia (sum of all cluster inertias)
    total_inertia = sum(inertia_results.values())
    
    return inertia_results, total_inertia

# 6. Compare Inertia Before/After
def compare_inertia(initial_emb, learned_emb, labels):
    """Compare inertia metrics between initial and learned embeddings"""
    initial_inertia, initial_total = calculate_inertia(initial_emb, labels)
    learned_inertia, learned_total = calculate_inertia(learned_emb, labels)
    
    # Create comparison DataFrame
    import pandas as pd
    comparison = pd.DataFrame({
        'word': list(initial_inertia.keys()),
        'Initial_Inertia': list(initial_inertia.values()),
        'Learned_Inertia': [learned_inertia[k] for k in initial_inertia.keys()],
        'Inertia_Reduction': [initial_inertia[k] - learned_inertia[k] for k in initial_inertia.keys()]
    })
    
    # Calculate percentage reduction
    comparison['Reduction_Pct'] = (comparison['Inertia_Reduction'] / comparison['Initial_Inertia']) * 100
    
    print(f"Total Inertia - Initial: {initial_total:.2f}, Learned: {learned_total:.2f}")
    print(f"Overall Reduction: {initial_total - learned_total:.2f} ({((initial_total - learned_total)/initial_total)*100:.1f}%)")
    
    return comparison.sort_values('Reduction_Pct', ascending=False)

# 7. Full Analysis Pipeline
def analyze_acoustic_space(input_folder, graph_file, model_path):
    """Complete analysis pipeline"""
    # 1. Load graph
    hetero_graph = load_heterograph(input_folder, graph_file)
    
    # 2. Get initial embeddings
    features, phoneme_labels = get_initial_embeddings(hetero_graph)
    initial_embeddings = features['acoustic']
    
    # 3. Visualize initial space
    print("\nVisualizing Initial Acoustic Space:")
    plot_embeddings(initial_embeddings, phoneme_labels, "Initial Acoustic Embeddings")
    
    # 4. Load model and get learned embeddings
    in_feats = {ntype: features[ntype].shape[1] for ntype in features}
    learned_embeddings = load_model_and_get_embeddings(model_path, hetero_graph, in_feats)
    
    # 5. Visualize learned space
    print("\nVisualizing Learned Acoustic Space:")
    plot_embeddings(learned_embeddings, phoneme_labels, "Learned Acoustic Embeddings")
    
    # 6. Compare inertia
    if phoneme_labels is not None:
        print("\nInertia Comparison:")
        inertia_comparison = compare_inertia(initial_embeddings, learned_embeddings, phoneme_labels)
        display(inertia_comparison.head(10))
        
        # Plot top improved clusters
        top_improved = inertia_comparison.nlargest(5, 'Reduction_Pct')
        plt.figure(figsize=(10, 6))
        sns.barplot(x='word', y='Reduction_Pct', data=top_improved)
        plt.title("Top word by Inertia Reduction")
        plt.ylabel("Inertia Reduction (%)")
        plt.show()
    else:
        print("No phoneme labels available for inertia calculation")
    
    return initial_embeddings, learned_embeddings, phoneme_labels

```

```python
import torch
import numpy as np
from sklearn.neighbors import NearestNeighbors
from collections import defaultdict, Counter
import matplotlib.pyplot as plt

def evaluate_cross_space_alignment(
    word_vectors,        # Tensor [n_mots, dim]
    acoustic_vectors,    # Tensor [n_acoustic, dim]
    word_labels,        # Liste des mots ['chat', 'chien', ...]
    acoustic_word_ids,  # Liste des IDs de mots [1, 3, 2, ...] correspondant aux labels acoustiques
    k=5,               # Nombre de voisins
    test_words=None     # Sous-ensemble de mots  tester (None = tous)
):
    """
    value l'alignement entre espaces linguistique et acoustique.
    
    Args:
        word_vectors: Embeddings des mots (Tensor)
        acoustic_vectors: Embeddings acoustiques (Tensor)
        word_labels: Noms des mots (liste)
        acoustic_word_ids: IDs des mots correspondants aux embeddings acoustiques
        k: Nombre de voisins  considrer
        test_words: Liste des mots  tester (None pour tous)
    
    Returns:
        Tuple: (rsultats dtaills, statistiques globales)
    """
    # Conversion des tenseurs en numpy si ncessaire
    word_vectors = word_vectors.cpu().numpy()
    acoustic_vectors = acoustic_vectors.cpu().numpy()
    
    # Cration d'un mapping mot_id -> indices acoustiques
    word_id_to_acoustic_indices = defaultdict(list)
    for acoustic_idx, word_id in enumerate(acoustic_word_ids):
        word_id_to_acoustic_indices[word_id].append(acoustic_idx)
    
    # Cration d'un mapping mot -> ID
    word_to_id = {word: idx for idx, word in enumerate(word_labels)}
    
    # Initialisation des modles de voisinage
    word_nbrs = NearestNeighbors(n_neighbors=k+1, metric='cosine').fit(word_vectors)
    acoustic_nbrs = NearestNeighbors(n_neighbors=k*3, metric='cosine').fit(acoustic_vectors)
    
    # Slection des mots  tester
    if test_words is None:
        test_words = word_labels
    else:
        test_words = [w for w in test_words if w in word_labels]
    
    results = {}
    
    for word in test_words:
        word_id = word_to_id[word]
        
        # Voisins linguistiques
        _, word_neighbor_idxs = word_nbrs.kneighbors([word_vectors[word_id]])
        linguistic_neighbors = [word_labels[i] for i in word_neighbor_idxs[0][1:k+1]]  # Exclure le mot lui-mme
        
        # Voisins acoustiques (agrgation sur tous les segments acoustiques du mot)
        acoustic_indices = word_id_to_acoustic_indices.get(word_id, [])
        all_acoustic_neighbors = []
        
        
        _, a_neighbor_idxs = acoustic_nbrs.kneighbors([word_vectors[word_id]])
            # Convertir les IDs acoustiques en mots
        neighbors = [word_labels[acoustic_word_ids[i]] for i in a_neighbor_idxs[0] 
                        if acoustic_word_ids[i] != word_id]  # Exclure le mot lui-mme
        all_acoustic_neighbors.extend(neighbors)
        
        # Prendre les k voisins acoustiques les plus frquents
        neighbor_counts = Counter(all_acoustic_neighbors)
        acoustic_neighbors = [w for w, _ in neighbor_counts.most_common(k)]
        
        # Calcul du recouvrement
        overlap = len(set(linguistic_neighbors) & set(acoustic_neighbors))
        
        # Stockage des rsultats
        results[word] = {
            'linguistic_neighbors': linguistic_neighbors,
            'acoustic_neighbors': acoustic_neighbors,
            'overlap': overlap,
            'acoustic_segments': len(acoustic_indices)
        }
    
    # Calcul des statistiques globales
    stats = {
        'mean_overlap': np.mean([r['overlap'] for r in results.values()]),
        'perfect_match_ratio': sum(1 for r in results.values() if r['overlap'] == k) / len(results),
        'words_tested': len(results),
        'mean_acoustic_segments': np.mean([r['acoustic_segments'] for r in results.values()])
    }
    
    return results, stats

# ----------------------------------------------------------
# Exemple d'utilisation
# ----------------------------------------------------------
# Donnes d'exemple ( remplacer par vos donnes relles)
n_words = 100
dim = 256

# Espace linguistique
word_vectors = torch.randn(n_words, dim)
word_labels = [f"word_{i}" for i in range(n_words)]

# Espace acoustique (plus d'instances que de mots)
n_acoustic = 500
acoustic_vectors = torch.randn(n_acoustic, dim)
acoustic_word_ids = torch.randint(0, n_words, (n_acoustic,)).tolist()  # Chaque segment acoustique est associ  un mot

# Appel de la fonction
results, stats = evaluate_cross_space_alignment(
    word_vectors=word_vectors,
    acoustic_vectors=acoustic_vectors,
    word_labels=word_labels,
    acoustic_word_ids=acoustic_word_ids,
    k=3,
    test_words=["word_10", "word_42", "word_75"]  # Optionnel: tester seulement certains mots
)

# Affichage des rsultats
print("=== Statistiques Globales ===")
print(f"Mots tests: {stats['words_tested']}")
print(f"Recouvrement moyen: {stats['mean_overlap']:.2f}/{k}")
print(f"Matchs parfaits: {stats['perfect_match_ratio']*100:.1f}%")
print(f"Segments acoustiques moyens par mot: {stats['mean_acoustic_segments']:.1f}")

print("\n=== Rsultats Dtails ===")
for word, data in results.items():
    print(f"\nMot: {word} ({data['acoustic_segments']} segments)")
    print(f"Voisins linguistiques: {data['linguistic_neighbors']}")
    print(f"Voisins acoustiques: {data['acoustic_neighbors']}")
    print(f"Recouvrement: {data['overlap']}/{k}")

# Visualisation
plt.figure(figsize=(10, 6))
overlaps = [r['overlap'] for r in results.values()]
plt.hist(overlaps, bins=np.arange(-0.5, k+1.5, 1), rwidth=0.8)
plt.xticks(range(k+1))
plt.title("Distribution du Recouvrement des Voisinages")
plt.xlabel("Nombre de voisins communs")
plt.ylabel("Nombre de mots")
plt.show()
```

```python
import numpy as np
import pickle
from sentence_transformers import SentenceTransformer  # Pour LaBSE
import torch

def generate_external_word_embeddings(word_list, embedding_method='labse', output_path="external_word_embeddings.pkl"):
    """
    Generate embeddings for external words using specified method and save in required format
    
    Args:
        word_list: List of words to generate embeddings for
        embedding_method: One of ['labse', 'fasttext', 'phonetic', 'articulatory']
        output_path: Path to save the output pickle file
    """
    # Vrification des mthodes supportes
    supported_methods = ['labse', 'fasttext', 'phonetic', 'articulatory']
    if embedding_method not in supported_methods:
        raise ValueError(f"Unsupported method. Choose from {supported_methods}")
    
    print(f"Generating {embedding_method} embeddings for {len(word_list)} words...")
    
    # Gnration des embeddings selon la mthode choisie
    if embedding_method == 'labse':
        # Initialisation du modle LaBSE
        model = SentenceTransformer('sentence-transformers/LaBSE')
        embeddings = model.encode(word_list, show_progress_bar=True)
        
    elif embedding_method == 'fasttext':
        # Ici vous devriez charger votre modle FastText pr-entran
        # Exemple simplifi (vous devrez adapter avec votre vrai modle)
        import fasttext
        model = fasttext.load_model("cc.en.300.bin")  # Remplacez par votre chemin
        embeddings = np.array([model.get_word_vector(word) for word in word_list])
        
    elif embedding_method == 'phonetic':
        # Mthode phontique - exemple simplifi
        from phonemizer import phonemize
        from collections import defaultdict
        
        # Phonmisation des mots
        phonemes = phonemize(word_list, language='en-us', backend='espeak')
        
        # Cration d'un vocabulaire phontique
        all_phonemes = set()
        for ph in phonemes:
            all_phonemes.update(ph.split())
        
        phon_to_idx = {ph: i for i, ph in enumerate(all_phonemes)}
        
        # Reprsentation one-hot (simplifie)
        embeddings = np.zeros((len(word_list), len(phon_to_idx)))
        for i, ph_seq in enumerate(phonemes):
            for ph in ph_seq.split():
                if ph in phon_to_idx:
                    embeddings[i, phon_to_idx[ph]] += 1
        
    elif embedding_method == 'articulatory':
        # Mthode articulatoire - exemple simplifi
        # Vous devrez implmenter votre propre logique ici
        articulatory_features = {
            # Exemple de caractristiques articulatoires
            'p': [1, 0, 0, 1, 0], 't': [0, 1, 0, 1, 0], 
            'k': [0, 0, 1, 1, 0], 'a': [0, 0, 0, 0, 1],
            # ... compltez avec vos propres caractristiques
        }
        
        embeddings = []
        for word in word_list:
            # Reprsentation moyenne des caractristiques articulatoires
            # (Ceci est un exemple trs simplifi)
            word_feats = np.zeros(5)  # Adaptez  votre dimension
            count = 0
            for letter in word.lower():
                if letter in articulatory_features:
                    word_feats += articulatory_features[letter]
                    count += 1
            if count > 0:
                word_feats /= count
            embeddings.append(word_feats)
        
        embeddings = np.array(embeddings)
    
    # Vrification de la forme des embeddings
    print(f"Generated embeddings shape: {embeddings.shape}")
    
    # Cration du dictionnaire de sortie
    output_data = {
        'vectors': embeddings,
        'labels': word_list,
        'embedding_method': embedding_method
    }
    
    # Sauvegarde dans un fichier pickle
    with open(output_path, 'wb') as f:
        pickle.dump(output_data, f)
    
    print(f"Embeddings saved to {output_path}")
    return output_data

# Exemple d'utilisation
if __name__ == "__main__":
    # Liste de mots externes (exemple)
    with open(f'subset_label_names_google_commands.pkl', 'rb') as f:
        label_ex_names = pickle.load(f)
    ex_label_names = list(label_ex_names.keys()) 
    external_words = ex_label_names
    
    
    # Gnration des embeddings avec LaBSE (par dfaut)
    labse_data = generate_external_word_embeddings(
        word_list=external_words,
        embedding_method='labse',
        output_path="external_word_embeddings_labse.pkl"
    )
    
    # Pour utiliser une autre mthode (ex: FastText)
    # fasttext_data = generate_external_word_embeddings(
    #     word_list=external_words,
    #     embedding_method='fasttext',
    #     output_path="external_word_embeddings_fasttext.pkl"
    # )
```

```python
#!/usr/bin/env python
# coding: utf-8

import torch
import dgl
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import networkx as nx
from collections import defaultdict, Counter
import os
from datetime import datetime
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows
from gnn_heto_link_pred_model import HeteroLinkGCN

class AlignmentEvaluator:
    def __init__(self, graph_path, model_path=None, label_names_path=None):
        """
        Initialize the evaluator with paths to required files
        
        Args:
            graph_path: Path to the heterograph file
            model_path: Path to the trained model (optional)
            label_names_path: Path to the label names pickle file (optional)
        """
        self.graph_path = graph_path
        self.model_path = model_path
        self.label_names_path = label_names_path
        self.hetero_graph = None
        self.word_labels = None
        self.results = {}
        
        # Set up plotting style
        plt.style.use('seaborn')
        sns.set_palette("husl")
    
    def load_data(self):
        """Load all required data (graph, labels, etc.)"""
        # Load heterograph
        glist, _ = dgl.load_graphs(self.graph_path)
        self.hetero_graph = glist[0]
        
        # Load label names if provided
        if self.label_names_path:
            with open(self.label_names_path, 'rb') as f:
                label_dic_names = pickle.load(f)
            self.word_labels = list(label_dic_names.keys())
        
        # Print graph characteristics
        self._print_graph_info()
    
    def _print_graph_info(self):
        """Print information about the heterograph"""
        print("\n=== Graph Characteristics ===")
        print("Node types:", self.hetero_graph.ntypes)
        print("Number of acoustic nodes:", self.hetero_graph.num_nodes('acoustic'))
        print("Number of word nodes:", self.hetero_graph.num_nodes('word'))
        print("Edge types:", self.hetero_graph.etypes)
        
        # Count edges
        for etype in self.hetero_graph.etypes:
            print(f"Number of '{etype}' edges:", self.hetero_graph.num_edges(etype))
        
        # Store graph info in results
        self.results['graph_info'] = {
            'node_types': self.hetero_graph.ntypes,
            'num_acoustic_nodes': self.hetero_graph.num_nodes('acoustic'),
            'num_word_nodes': self.hetero_graph.num_nodes('word'),
            'edge_types': self.hetero_graph.etypes,
            'num_edges': {etype: self.hetero_graph.num_edges(etype) 
                         for etype in self.hetero_graph.etypes}
        }
    
    def get_embeddings(self):
        """Extract initial and learned embeddings"""
        # Get initial features
        features = {
            'acoustic': self.hetero_graph.nodes['acoustic'].data['feat'],
            'word': self.hetero_graph.nodes['word'].data['feat']
        }
        phoneme_labels = self.hetero_graph.nodes['acoustic'].data.get('label', None)
        
        # Get learned embeddings if model is provided
        learned_embeddings = None
        if self.model_path:
            in_feats = {ntype: features[ntype].shape[1] for ntype in features}
            hidden_size = 512
            linear_hidden_size = 64
            nb_phon = features['word'].shape[1]
            
            model = HeteroLinkGCN(in_feats, hidden_size, nb_phon, linear_hidden_size)
            model.load_state_dict(torch.load(self.model_path))
            model.eval()
            
            with torch.no_grad():
                learned_embeddings = model(self.hetero_graph, features)['acoustic']
        
        return features['acoustic'], learned_embeddings, phoneme_labels, features['word']
    
    def calculate_inertia(self, embeddings, labels):
        """Calculate inertia for phoneme clusters"""
        embeddings_np = embeddings.cpu().numpy()
        labels_np = labels.cpu().numpy()
        
        unique_labels = np.unique(labels_np)
        inertia_results = {}
        
        for label in unique_labels:
            cluster_points = embeddings_np[labels_np == label]
            if len(cluster_points) > 1:
                centroid = np.mean(cluster_points, axis=0)
                distances = np.sum((cluster_points - centroid) ** 2, axis=1)
                inertia = np.sum(distances)
                inertia_results[label] = inertia
            else:
                inertia_results[label] = 0.0
        
        total_inertia = sum(inertia_results.values())
        return inertia_results, total_inertia
    
    def evaluate_inertia_reduction(self):
        """Evaluate the reduction in inertia from initial to learned embeddings"""
        initial_emb, learned_emb, labels, _ = self.get_embeddings()
        
        if learned_emb is None:
            print("No model provided - skipping inertia evaluation")
            return None
        
        initial_inertia, initial_total = self.calculate_inertia(initial_emb, labels)
        learned_inertia, learned_total = self.calculate_inertia(learned_emb, labels)
        
        # Create comparison DataFrame
        comparison = pd.DataFrame({
            'word': list(initial_inertia.keys()),
            'initial_inertia': list(initial_inertia.values()),
            'learned_inertia': [learned_inertia[k] for k in initial_inertia.keys()],
            'inertia_reduction': [initial_inertia[k] - learned_inertia[k] 
                                 for k in initial_inertia.keys()]
        })
        
        comparison['reduction_pct'] = (comparison['inertia_reduction'] / 
                                     comparison['initial_inertia']) * 100
        
        # Store results
        self.results['inertia'] = {
            'initial_total': initial_total,
            'learned_total': learned_total,
            'total_reduction': initial_total - learned_total,
            'reduction_pct': ((initial_total - learned_total) / initial_total) * 100,
            'per_word': comparison
        }
        
        print("\n=== Inertia Reduction Results ===")
        print(f"Total Inertia - Initial: {initial_total:.2f}, Learned: {learned_total:.2f}")
        print(f"Overall Reduction: {initial_total - learned_total:.2f} "
              f"({((initial_total - learned_total)/initial_total)*100:.1f}%)")
        
        return comparison
    
    def evaluate_cross_space_alignment(self, k=5, test_words=None):
        """Evaluate cross-space alignment between word and acoustic embeddings"""
        _, _, acoustic_labels, word_emb = self.get_embeddings()
        acoustic_emb = self.hetero_graph.nodes['acoustic'].data['feat']
        
        if self.word_labels is None:
            print("No word labels provided - skipping cross-space evaluation")
            return None, None
        
        # Convert to numpy
        word_vectors = word_emb.cpu().numpy()
        acoustic_vectors = acoustic_emb.cpu().numpy()
        acoustic_word_ids = acoustic_labels.cpu().numpy()
        
        # Create mappings
        word_id_to_acoustic_indices = defaultdict(list)
        for acoustic_idx, word_id in enumerate(acoustic_word_ids):
            word_id_to_acoustic_indices[word_id].append(acoustic_idx)
        
        word_to_id = {word: idx for idx, word in enumerate(self.word_labels)}
        
        # Initialize nearest neighbors models
        word_nbrs = NearestNeighbors(n_neighbors=k+1, metric='cosine').fit(word_vectors)
        acoustic_nbrs = NearestNeighbors(n_neighbors=k*3, metric='cosine').fit(acoustic_vectors)
        
        # Select words to test
        if test_words is None:
            test_words = self.word_labels
        else:
            test_words = [w for w in test_words if w in self.word_labels]
        
        results = {}
        
        for word in test_words:
            word_id = word_to_id[word]
            
            # Linguistic neighbors
            _, word_neighbor_idxs = word_nbrs.kneighbors([word_vectors[word_id]])
            linguistic_neighbors = [self.word_labels[i] for i in word_neighbor_idxs[0][1:k+1]]
            
            # Acoustic neighbors
            acoustic_indices = word_id_to_acoustic_indices.get(word_id, [])
            all_acoustic_neighbors = []
            
            for idx in acoustic_indices:
                _, a_neighbor_idxs = acoustic_nbrs.kneighbors([acoustic_vectors[idx]])
                neighbors = [self.word_labels[acoustic_word_ids[i]] for i in a_neighbor_idxs[0] 
                            if acoustic_word_ids[i] != word_id]
                all_acoustic_neighbors.extend(neighbors[:k])  # Take top k for each segment
            
            # Get most common acoustic neighbors
            neighbor_counts = Counter(all_acoustic_neighbors)
            acoustic_neighbors = [w for w, _ in neighbor_counts.most_common(k)]
            
            # Calculate overlap
            overlap = len(set(linguistic_neighbors) & set(acoustic_neighbors))
            
            results[word] = {
                'linguistic_neighbors': linguistic_neighbors,
                'acoustic_neighbors': acoustic_neighbors,
                'overlap': overlap,
                'acoustic_segments': len(acoustic_indices)
            }
        
        # Calculate global statistics
        stats = {
            'mean_overlap': np.mean([r['overlap'] for r in results.values()]),
            'perfect_match_ratio': sum(1 for r in results.values() if r['overlap'] == k) / len(results),
            'words_tested': len(results),
            'mean_acoustic_segments': np.mean([r['acoustic_segments'] for r in results.values()])
        }
        
        # Store results
        self.results['cross_space'] = {
            'per_word': results,
            'stats': stats,
            'k': k
        }
        
        print("\n=== Cross-Space Alignment Results ===")
        print(f"Words tested: {stats['words_tested']}")
        print(f"Mean overlap: {stats['mean_overlap']:.2f}/{k}")
        print(f"Perfect matches: {stats['perfect_match_ratio']*100:.1f}%")
        print(f"Mean acoustic segments per word: {stats['mean_acoustic_segments']:.1f}")
        
        return results, stats
    
    def visualize_results(self, output_dir=None):
        """Generate visualizations of the results"""
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)
        
        # Inertia reduction visualization
        if 'inertia' in self.results:
            inertia_df = self.results['inertia']['per_word']
            
            # Top improved clusters
            top_improved = inertia_df.nlargest(10, 'reduction_pct')
            plt.figure(figsize=(12, 6))
            sns.barplot(x='word', y='reduction_pct', data=top_improved)
            plt.title("Top Words by Inertia Reduction Percentage")
            plt.ylabel("Inertia Reduction (%)")
            plt.xticks(rotation=45)
            plt.tight_layout()
            
            if output_dir:
                plt.savefig(os.path.join(output_dir, 'top_inertia_reduction.png'))
            plt.show()
            
            # Initial vs learned inertia scatter plot
            plt.figure(figsize=(8, 8))
            sns.scatterplot(x='initial_inertia', y='learned_inertia', data=inertia_df)
            plt.plot([0, inertia_df['initial_inertia'].max()], 
                    [0, inertia_df['initial_inertia'].max()], 'r--')
            plt.title("Initial vs Learned Inertia")
            plt.xlabel("Initial Inertia")
            plt.ylabel("Learned Inertia")
            
            if output_dir:
                plt.savefig(os.path.join(output_dir, 'inertia_comparison.png'))
            plt.show()
        
        # Cross-space alignment visualization
        if 'cross_space' in self.results:
            k = self.results['cross_space']['k']
            overlaps = [r['overlap'] for r in self.results['cross_space']['per_word'].values()]
            
            plt.figure(figsize=(10, 6))
            plt.hist(overlaps, bins=np.arange(-0.5, k+1.5, 1), rwidth=0.8)
            plt.xticks(range(k+1))
            plt.title("Distribution of Neighbor Overlap Between Spaces")
            plt.xlabel("Number of common neighbors")
            plt.ylabel("Number of words")
            
            if output_dir:
                plt.savefig(os.path.join(output_dir, 'neighbor_overlap_distribution.png'))
            plt.show()
    
    def evaluate_external_words_alignment(self, external_word_vectors, external_word_labels, k=5):
        """
        Evaluate alignment for words not in the original graph using their external embeddings
        
        Args:
            external_word_vectors: Tensor or numpy array of shape [n_words, dim]
            external_word_labels: List of word strings
            k: Number of neighbors to consider
        """
        # Get acoustic embeddings and labels from graph
        acoustic_emb = self.hetero_graph.nodes['acoustic'].data['feat'].cpu().numpy()
        acoustic_word_ids = self.hetero_graph.nodes['acoustic'].data['label'].cpu().numpy()
        
        # Convert word labels to IDs if we have them in the graph
        if self.word_labels:
            word_to_id = {word: idx for idx, word in enumerate(self.word_labels)}
        
        # Initialize nearest neighbors models
        acoustic_nbrs = NearestNeighbors(n_neighbors=k*3, metric='cosine').fit(acoustic_emb)
        
        results = {}
        
        for word, vector in zip(external_word_labels, external_word_vectors):
            # Find linguistic neighbors (using external embeddings)
            linguistic_distances = pairwise_distances(
                [vector], external_word_vectors, metric='cosine')[0]
            sorted_indices = np.argsort(linguistic_distances)
            linguistic_neighbors = [external_word_labels[i] for i in sorted_indices[1:k+1]]  # Exclude self
            
            # Find acoustic neighbors
            _, a_neighbor_idxs = acoustic_nbrs.kneighbors([vector])
            acoustic_neighbors = []
            
            for idx in a_neighbor_idxs[0]:
                word_id = acoustic_word_ids[idx]
                if self.word_labels and word_id < len(self.word_labels):
                    neighbor_word = self.word_labels[word_id]
                    if neighbor_word != word:  # Exclude self if word exists in graph
                        acoustic_neighbors.append(neighbor_word)
                else:
                    # If we don't have word labels, just use the ID
                    acoustic_neighbors.append(f"word_{word_id}")
            
            # Take top k unique acoustic neighbors
            acoustic_neighbors = list(dict.fromkeys(acoustic_neighbors))[:k]
            
            # Calculate overlap
            overlap = len(set(linguistic_neighbors) & set(acoustic_neighbors))
            
            results[word] = {
                'linguistic_neighbors': linguistic_neighbors,
                'acoustic_neighbors': acoustic_neighbors,
                'overlap': overlap
            }
        
        # Calculate global statistics
        stats = {
            'mean_overlap': np.mean([r['overlap'] for r in results.values()]),
            'perfect_match_ratio': sum(1 for r in results.values() if r['overlap'] == k) / len(results),
            'words_tested': len(results)
        }
        
        # Store results
        self.results['external_words'] = {
            'per_word': results,
            'stats': stats,
            'k': k
        }
        
        print("\n=== External Words Alignment Results ===")
        print(f"Words tested: {stats['words_tested']}")
        print(f"Mean overlap: {stats['mean_overlap']:.2f}/{k}")
        print(f"Perfect matches: {stats['perfect_match_ratio']*100:.1f}%")
        
        return results, stats

    def save_to_excel(self, output_path):
        """Save all results to an Excel file"""
        # Create a new workbook
        wb = Workbook()
        
        # Add graph info sheet
        ws = wb.active
        ws.title = "Graph Info"
        
        graph_info = self.results.get('graph_info', {})
        
        # Convert lists to strings for Excel compatibility
        excel_compatible_info = {}
        for key, value in graph_info.items():
            if isinstance(value, (list, dict)):
                excel_compatible_info[key] = str(value)
            else:
                excel_compatible_info[key] = value
        
        # Create DataFrame and write to sheet
        df = pd.DataFrame([excel_compatible_info])
        for row in dataframe_to_rows(df, index=False, header=True):
            ws.append(row)
        
        # Add inertia results sheet
        if 'inertia' in self.results:
            ws = wb.create_sheet("Inertia Results")
            
            # Summary stats
            summary_data = {
                'initial_total_inertia': [self.results['inertia']['initial_total']],
                'learned_total_inertia': [self.results['inertia']['learned_total']],
                'total_reduction': [self.results['inertia']['total_reduction']],
                'reduction_percentage': [self.results['inertia']['reduction_pct']]
            }
            
            for row in dataframe_to_rows(pd.DataFrame(summary_data), index=False, header=True):
                ws.append(row)
            
            # Per-word results
            ws.append([])  # Empty row
            ws.append(["Per-Word Results"])
            
            for row in dataframe_to_rows(self.results['inertia']['per_word'], index=False, header=True):
                ws.append(row)
        
        # Add cross-space alignment sheet
        if 'cross_space' in self.results:
            ws = wb.create_sheet("Cross-Space Alignment")
            
            # Summary stats
            stats = self.results['cross_space']['stats']
            summary_data = {
                'k': [self.results['cross_space']['k']],
                'words_tested': [stats['words_tested']],
                'mean_overlap': [stats['mean_overlap']],
                'perfect_match_ratio': [stats['perfect_match_ratio']],
                'mean_acoustic_segments': [stats['mean_acoustic_segments']]
            }
            
            for row in dataframe_to_rows(pd.DataFrame(summary_data), index=False, header=True):
                ws.append(row)
        
        # Add external words alignment sheet
        if 'external_words' in self.results:
            ws = wb.create_sheet("External Words Alignment")
            
            # Summary stats
            stats = self.results['external_words']['stats']
            summary_data = {
                'k': [self.results['external_words']['k']],
                'words_tested': [stats['words_tested']],
                'mean_overlap': [stats['mean_overlap']],
                'perfect_match_ratio': [stats['perfect_match_ratio']]
            }
            
            for row in dataframe_to_rows(pd.DataFrame(summary_data), index=False, header=True):
                ws.append(row)
            
            # Per-word results
            ws.append([])
            ws.append(["Per-Word Results"])
            
            simplified_results = []
            for word, data in self.results['external_words']['per_word'].items():
                simplified_results.append({
                    'word': word,
                    'overlap': data['overlap'],
                    'linguistic_neighbors': str(data['linguistic_neighbors']),
                    'acoustic_neighbors': str(data['acoustic_neighbors'])
                })
            
            for row in dataframe_to_rows(pd.DataFrame(simplified_results), index=False, header=True):
                ws.append(row)
        
        # Save the workbook
        wb.save(output_path)
        print(f"\nResults saved to {output_path}")

```

```python
def main():
    # Configuration
    config = {
        'graph_path': "saved_graphs/timit/clique/filter/fixed/labse/hetero_graph_0_1_0_7000.dgl",
        'model_path': "models/hetero_gnn_edge_regressor.pth",
        'label_names_path': "subset_label_names_timit.pkl",
        'output_excel': f"alignment_results_{datetime.now().strftime('%Y%m%d_%H%M%S')}.xlsx",
        'output_dir': "results_visualizations",
        'external_words_path': "external_word_embeddings_labse.pkl"  # Nouveau chemin pour les mots externes
    }
    
    # Initialize evaluator
    evaluator = AlignmentEvaluator(
        graph_path=config['graph_path'],
        model_path=config['model_path'],
        label_names_path=config['label_names_path']
    )
    
    # Load data
    evaluator.load_data()
    
    # Run standard evaluations
    evaluator.evaluate_inertia_reduction()
    evaluator.evaluate_cross_space_alignment(k=3)
    
    # Evaluate external words if provided
    if os.path.exists(config['external_words_path']):
        with open(config['external_words_path'], 'rb') as f:
            external_data = pickle.load(f)
            external_word_vectors = external_data['vectors']
            external_word_labels = external_data['labels']
        
        evaluator.evaluate_external_words_alignment(
            external_word_vectors=external_word_vectors,
            external_word_labels=external_word_labels,
            k=3
        )
    
    # Visualize and save results
    evaluator.visualize_results(output_dir=config['output_dir'])
    evaluator.save_to_excel(output_path=config['output_excel'])

if __name__ == "__main__":
    main()
```
